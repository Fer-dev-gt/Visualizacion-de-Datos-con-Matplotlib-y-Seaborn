```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
print(x)
print(y)
```

```python
"""
This code creates a figure with a single subplot and plots the data points (x, y) in red color.


Parameters:
- x: array-like or scalar. The x-coordinates of the data points.
- y: array-like or scalar. The y-coordinates of the data points.
"""

fig, axes = plt.subplots()
axes.plot(x, y, 'r')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')
```

```python
# Otra forma de hacer lo mismo con diferente sintaxis
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'r')
ax2.plot(y, x, 'b')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.cos(x), 'b')
axes[0, 1].plot(x, np.sin(x), 'r')
axes[0, 2].plot(x, np.tan(x), 'k')
axes[0, 3].plot(x, np.tanh(x), 'y')
axes[1, 0].plot(x, np.arccos(x), 'b')
axes[1, 1].plot(x, np.arcsin(x), 'g')
axes[1, 2].plot(x, np.arctan(x), 'r')
axes[1, 3].plot(x, np.cos(x) ** 2)

```

```python
# Otra forma de hacer lo mismo con diferente sintaxis
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x), 'bo')
ax2.plot(x, np.sin(x), 'r>')
ax3.plot(x, np.tan(x), 'ko')
ax4.plot(x, np.tanh(x), 'y1')
ax5.plot(x, np.arccos(x), 'bs')
ax6.plot(x, np.arcsin(x), 'gp')
ax7.plot(x, np.arctan(x), 'r:')
ax8.plot(x, np.cos(x) ** 2, 'r--')
fig.tight_layout()
```
