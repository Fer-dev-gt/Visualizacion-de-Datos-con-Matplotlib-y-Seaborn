```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
print(x)
print(y)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].plot(x, y, 'r', label='$sin(x)$')
axes[0].legend()
axes[0].set_title('Relacin entre x - y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(y, x, 'g')
axes[1].set_title('Relacin entre y - x')
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
```

```python
# Forma de graficar con plt
plt.plot(x, y, label='$sin(x)$')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Graficando los armonicos de una onda
y = np.zeros(100)
x = np.linspace(0, 20, 100)
n = np.linspace(1, 9, 9, endpoint=True)

harmonics = []

for idx, element in enumerate(n):
    y = np.sin(element * x)
    harmonics.append(y)

harmonics = np.array(harmonics)
harmonics = np.expand_dims(harmonics, axis=0)
harmonics = harmonics.reshape((3, 3, 100))

nrows = 3
ncols = 3

fig, axes = plt.subplots(nrows, ncols, figsize=(15, 15))

for idx in range(nrows):
    for jdx in range(ncols):
        axes[idx, jdx].plot(x, harmonics[idx, jdx, :], label=f'$sin({n[idx]})$')
        axes[idx, jdx].legend(loc='upper left', bbox_to_anchor=(1, 1))

fig.tight_layout()

# Explicacion 
# Cada fila de la matriz representa un armonico
# Cada columna representa un punto en el tiempo
# Cada celda representa el valor de la onda en ese punto en el tiempo
# Cada grfico representa un armonico de la onda
# Cada grfico tiene un color diferente
# Cada grfico tiene una etiqueta diferente
# Cada grfico tiene una leyenda
# Cada grfico tiene un titulo
# Cada grfico tiene un eje x
# Cada grfico tiene un eje y
'''
The first part of the code is setting up the x-axis values and the frequencies for the sine waves. The np.zeros(100) function creates an array of 100 zeros, which is stored in the variable y. The np.linspace(0, 20, 100) function generates 100 evenly spaced numbers between 0 and 20, which are stored in the variable x. The np.linspace(1, 9, 9, endpoint=True) function generates 9 evenly spaced numbers between 1 and 9, which are stored in the variable n.

The harmonics list is then created to store the y-values for each sine wave. The for loop with enumerate(n) iterates over each frequency in n, calculates the corresponding y-values using np.sin(element * x), and appends the result to the harmonics list.

The harmonics list is then converted to a numpy array and reshaped into a 3D array with shape (3, 3, 100) using the np.array() and reshape() functions. This is done to facilitate the plotting of the sine waves in a 3x3 grid.

The plt.subplots(nrows, ncols, figsize=(15, 15)) function is used to create a 3x3 grid of subplots with a figure size of 15x15. The nested for loops then iterate over each subplot in the grid, plot the corresponding sine wave using the plot() function, and add a legend using the legend() function.

Finally, fig.tight_layout() is called to adjust the subplot parameters so that the subplots fit into the figure area nicely. This is a common practice to prevent overlapping between subplots.
'''
```
