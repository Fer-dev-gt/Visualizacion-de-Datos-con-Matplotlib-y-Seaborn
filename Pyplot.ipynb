```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
print('Valores de lista X: ', x)
print('Valores de lista Y: ', y)
```

```python
def plot_data(x, y):
    """
    Plots the given data using Matplotlib's plot function.

    Parameters:
    x (array-like): The x-coordinates of the data points.
    y (array-like): The y-coordinates of the data points.

    Returns:
    None
    """
    plt.plot(x, y)
    plt.show()
plot_data(x, y)
```

```python
# Cambiando el color de la gr√°fica
plt.plot(x, y, 'r')
plt.show()
```

```python
# Graficando una gr√°fica de pastel
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
# Graficando una gr√°fica de Scatter con datos aleatorios
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
```
