```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['USA', 'Brazil', 'India', 'Russia', 'South Africa']
population = [331002651, 212559417, 1380004385, 145934462, 59308690]
```

```python
plt.bar(country, population, width=0.5, color=['green', 'blue'])
plt.xticks(np.arange(5), ('Estados Unidos', 'Brasil', 'India', 'Rusia', 'Sudafrica'), rotation=45)
plt.ylabel('Poblacion')
plt.title('Poblacion de los paises mas grandes')
plt.show()
```

```python
plt.barh(country, population, color=['green', 'blue', 'red', 'orange', 'purple'])
plt.yticks(np.arange(5), ('Estados Unidos', 'Brasil', 'India', 'Rusia', 'Sudafrica'))
plt.xlabel('Poblacion')
plt.title('Poblacion de los paises mas grandes')
plt.show()
```
