```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(0, 50, 100)
print(data)
```

```python
# Grafico de Histograma
plt.hist(data, bins=50, alpha=0.5, color='blue', histtype='bar')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.hist(data, bins=50, alpha=0.5, color='blue', histtype='step')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequencia')
plt.show()
```

```python
# Grafico de Boxplot
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.title('Boxplot')
plt.show()
```

```python
# Grafico de Scatter
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
```
