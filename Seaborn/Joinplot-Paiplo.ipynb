```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='hist')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='kde')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='hist', marginal_ticks=True)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=20, fill=True, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
# Para no tener graficos duplicados
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='line')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='scatter')
plt.show()
```
