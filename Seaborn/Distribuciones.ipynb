```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=50)
plt.show()
```

```python
# Misma grafica pero acumulan los datos
sns.histplot(data=tips, x='tip', bins=50, cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability', multiple='stack')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability', multiple='dodge')
plt.show()
```

```python
# Plots the kernel density estimate (KDE) of the tip distribution based on gender.

sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
# sns.displot nos sirve para no tener que aprendernos de memoria todas las opciones de los plots ya que con displot podemos hacerlo todo solo cambiando el parametro kind
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='ecdf')
plt.show()
```

```python
sns.displot(data=tips, x='smoker', hue='sex', kind='hist', stat='count')
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips['Weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.displot(data=tips, x='tip', hue='Weekend', kde=True, multiple='dodge')
```
