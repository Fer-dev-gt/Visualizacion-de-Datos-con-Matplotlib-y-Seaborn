```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
# Obtenemos la correlación entre las variables numéricas
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5)
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black')
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black', vmin=-1, vmax=1)
plt.show()
```
