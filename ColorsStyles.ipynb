```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
print(x)
```

```python
print(plt.style.available)
```

```python
plt.style.use('Solarize_Light2')
```

## Matlab style

```python
fig, ax = plt.subplots(figsize=(8, 8))   
ax.plot(x, x**2, label="y = x**2") 
ax.plot(x, x**3, 'r.-',label="y = x**3")
ax.plot(x, x**4, 'g:',label="y = x**4")
ax.legend() 
```

## PyPlot Style

```python
fig, ax = plt.subplots(figsize=(8, 8))   
ax.plot(x, x**2, color='blue', linewidth=2, linestyle='--', marker='o', markersize=10, label="y = x**2") 
ax.plot(x, x**3, color='red', alpha=0.2, linewidth=5, label="y = x**3", marker='P', markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='green')
ax.plot(x, x**4, color='#FFDD44', linewidth=3, linestyle='-.', label="y = x**4")
ax.legend() 
```

```python
# De esta forma podemos ver como se ven los diferentes estilos de matplotlib
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(5,5))
  ax.plot(x)
```
