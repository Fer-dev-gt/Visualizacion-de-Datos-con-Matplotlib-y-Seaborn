```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
print(x)
print(y)
```

```python
# Usando Subplots para mostrar varios graficos en una misma ventana
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.hist(y, bins=10, color='g')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.plot(y, x, 'b:')
plt.subplot(1, 2, 2)
plt.pie(y)
```

```python
# Mostrando 2 figuras en una misma columna
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 1, 2)
plt.plot(y, x, 'b')
```
